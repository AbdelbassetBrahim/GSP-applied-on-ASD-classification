```python
def plot_scatter (pvals_permut,Accuracy,marker, label):
    import matplotlib 
    
    Accuracy=Accuracy*100
    Accuracy_sig=np.argwhere(pvals_permut<1e-2)
    Accuracy_nonsig=np.argwhere(pvals_permut>1e-2)
    plt.scatter(Accuracy_sig,Accuracy[ Accuracy_sig],marker=marker,c="red",label=label)
    plt.scatter(Accuracy_nonsig,Accuracy[Accuracy_nonsig],marker=marker,c="blue") 
    
    ind = np.arange(len(Accuracy))
    k= np.arange(10,360,10)
    plt.xticks(ind,k,rotation='60')
    matplotlib.rc('xtick', labelsize=20) 
    
  
    
   
    plt.tight_layout()
   
   
```
